```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
import cv2
import os
import matplotlib.pyplot as plt
import seaborn as sns  # Added import
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
from tqdm.notebook import tqdm
import random

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
random.seed(42)

print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# 1. Data Loading and Preprocessing
def load_and_preprocess_data(img_size=(128, 128)):
    images = []
    labels = []

    # Define correct paths for Kaggle
    normal_path = "/kaggle/input/ct-kidney-dataset-normal-cyst-tumor-and-stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/Normal"
    stone_path = "/kaggle/input/ct-kidney-dataset-normal-cyst-tumor-and-stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/Stone"

    # Process Normal images
    print("\nProcessing Normal images...")
    files = os.listdir(normal_path)
    for img_name in tqdm(files):
        img_path = os.path.join(normal_path, img_name)
        try:
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, img_size)
            images.append(img)
            labels.append(0)  # 0 for Normal
        except Exception as e:
            print(f"Error loading image {img_path}: {str(e)}")

    # Process Stone images
    print("\nProcessing Stone images...")
    files = os.listdir(stone_path)
    for img_name in tqdm(files):
        img_path = os.path.join(stone_path, img_name)
        try:
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.resize(img, img_size)
            images.append(img)
            labels.append(1)  # 1 for Stone
        except Exception as e:
            print(f"Error loading image {img_path}: {str(e)}")

    images = np.array(images, dtype='float32') / 255.0  # Normalize pixel values
    labels = np.array(labels)

    return images, labels

# 2. Create CNN+LSTM Model
def create_cnn_lstm_model(input_shape):
    model = models.Sequential([
        # CNN layers
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape),
        layers.BatchNormalization(),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Reshape for LSTM
        layers.Reshape((16, -1)),

        # LSTM layers
        layers.LSTM(128, return_sequences=True),
        layers.Dropout(0.5),
        layers.LSTM(64),
        layers.Dropout(0.5),

        # Dense layers
        layers.Dense(256, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(1, activation='sigmoid')
    ])

    return model

# 3. Data Augmentation
def create_data_generator():
    return ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.15,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

# 4. Training Function
def train_model(X_train, X_test, y_train, y_test):
    model = create_cnn_lstm_model(input_shape=(128, 128, 3))

    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-4)

    model.compile(
        optimizer=optimizer,
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    model.summary()

    datagen = create_data_generator()
    datagen.fit(X_train)

    # Compute class weights
    from sklearn.utils.class_weight import compute_class_weight
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weights = dict(enumerate(class_weights))
    print(f"\nClass Weights: {class_weights}")

    # Callbacks
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=1e-7
        ),
        tf.keras.callbacks.ModelCheckpoint(
            '/kaggle/working/best_model.keras',
            monitor='val_loss',
            save_best_only=True
        )
    ]

    # Train the model
    history = model.fit(
        datagen.flow(X_train, y_train, batch_size=32),
        epochs=100,
        validation_data=(X_test, y_test),
        callbacks=callbacks,
        class_weight=class_weights
    )

    return model, history

# 5. Visualization Functions
def plot_training_history(history):
    # Use seaborn for styling
    sns.set_theme()  # Added this line

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 7))

    # Accuracy plot
    ax1.plot(history.history['accuracy'], label='Train Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Model Accuracy', pad=20, size=15)
    ax1.set_ylabel('Accuracy', size=12)
    ax1.set_xlabel('Epoch', size=12)
    ax1.legend(loc='lower right')
    ax1.grid(True)

    # Loss plot
    ax2.plot(history.history['loss'], label='Train Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Model Loss', pad=20, size=15)
    ax2.set_ylabel('Loss', size=12)
    ax2.set_xlabel('Epoch', size=12)
    ax2.legend(loc='upper right')
    ax2.grid(True)

    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png')
    plt.show()

# 6. Model Evaluation
def evaluate_model(model, X_test, y_test):
    # Evaluate model
    test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
    print(f"\nTest accuracy: {test_accuracy*100:.2f}%")

    # Make predictions
    y_pred = model.predict(X_test, verbose=1)
    y_pred_classes = (y_pred > 0.5).astype(int).flatten()

    # Classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred_classes, target_names=['Normal', 'Stone']))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred_classes)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Normal', 'Stone'],
                yticklabels=['Normal', 'Stone'])
    plt.title('Confusion Matrix', pad=20, size=15)
    plt.ylabel('True Label', size=12)
    plt.xlabel('Predicted Label', size=12)
    plt.savefig('/kaggle/working/confusion_matrix.png')
    plt.show()

# 7. Prediction Function
def predict_kidney_stone(image_path, model_path='/kaggle/working/best_model.keras'):
    # Load model
    model = tf.keras.models.load_model(model_path)

    # Preprocess image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Error: Unable to read image at {image_path}")
        return None, None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img, (128, 128))
    img_resized = img_resized / 255.0
    img_resized = np.expand_dims(img_resized, axis=0)

    # Make prediction
    prediction = model.predict(img_resized)
    probability = prediction[0][0]

    result = "Kidney Stone" if probability > 0.5 else "Normal"
    print(f"\nPrediction: {result}")
    print(f"Probability: {probability:.2f}")

    # Display the image
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.title(f'Prediction: {result}\nProbability: {probability:.2f}')
    plt.axis('off')
    plt.show()

    return result, probability

# 8. Main execution
def main():
    # Load and preprocess data
    print("\nLoading and preprocessing data...")
    images, labels = load_and_preprocess_data()

    # Check class distribution
    unique, counts = np.unique(labels, return_counts=True)
    class_distribution = dict(zip(unique, counts))
    print(f"\nClass distribution: {class_distribution}")  # 0: Normal, 1: Stone

    # Shuffle and split data
    from sklearn.utils import shuffle
    images, labels = shuffle(images, labels, random_state=42)

    X_train, X_test, y_train, y_test = train_test_split(
        images, labels, test_size=0.2, random_state=42, stratify=labels
    )

    # Print split statistics
    print(f"\nTraining set size: {len(X_train)}")
    print(f"Test set size: {len(X_test)}")

    # Train model
    print("\nTraining model...")
    model, history = train_model(X_train, X_test, y_train, y_test)

    # Plot training history
    plot_training_history(history)

    # Evaluate model
    print("\nEvaluating model...")
    evaluate_model(model, X_test, y_test)

    # Save the final model
    model.save('/kaggle/working/kidney_stone_detector_final.keras')
    print("\nFinal model saved to /kaggle/working/kidney_stone_detector_final.keras")

    # Generate predictions on some test images
    print("\nGenerating example predictions...")
    for idx in range(5):
        if idx < len(X_test):
            img = X_test[idx]
            true_label = 'Stone' if y_test[idx] == 1 else 'Normal'
            img_display = (img * 255).astype(np.uint8)
            prediction = model.predict(np.expand_dims(img, axis=0))[0][0]
            predicted_label = 'Stone' if prediction > 0.5 else 'Normal'
            plt.figure(figsize=(4, 4))
            plt.imshow(img_display)
            plt.title(f"True: {true_label}\nPredicted: {predicted_label}\nProbability: {prediction:.2f}")
            plt.axis('off')
            plt.savefig(f'/kaggle/working/example_prediction_{idx}.png')
            plt.show()

if __name__ == "__main__":
    main()
```
